```python
import matplotlib.pyplot as plt  
import pandas as pd
import numpy as np
from numpy import set_printoptions
from numpy import mean
# Models
from sklearn.model_selection import train_test_split

dataframe = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/dane_ml(1)/adam.csv", sep = ';')
```

```python
dataframe.head(5)
```

```python
signals =np.where(dataframe.FLAG==1)[0]
```

```python
numberSignals =int(len(signals)/2)
```

```python
max = max(signals[0:numberSignals])
```

```python
signals[0:numberSignals]
```

```python
fig = plt.figure(figsize=(12,6))
column = ["YAW","ROLL","PITCH"]
signalTimestamp = signals[0:numberSignals]
for i in column:
  plt.plot(dataframe.TIME[0:max], 
             dataframe[i][0:max], 
             label=i)
    
for s in signalTimestamp:
  plt.axvline(dataframe.TIME[s], color="red")
  #plt.xticks(np.arange(0.0, 1.1, step=0.1))
plt.xlabel("time", fontsize=15)

  #plt.yticks(np.arange(0.0, 1.1, step=0.1))
plt.ylabel("value", fontsize=15)

plt.title('Result', fontweight='bold', fontsize=15)
plt.legend(prop={'size':13}, loc='lower right')

plt.show()
```
